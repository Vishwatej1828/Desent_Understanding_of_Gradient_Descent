# Printing Optimized Line by Gradient Descent Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = [[1, 2.1, 3.5, 3.7, 4.17, 4.96, 5.1, 5.6, 6.1, 7, 7.9, 8.4, 9,9.2,10]]
y = [[1,5,5.2,5,4.5,9,10,11,11,12,15,14,17,20,19]]
x = np.array(x).reshape(15,1)
y = np.array(y).reshape(15,1)
```

```python
a = 0
b = 0

epoch =2000
learn = 0.001

for i in range(epoch):
    y_pred = a + (b*x)
    a = a -(-2/len(y))*learn*sum(y-y_pred)
    b = b - (-2/len(y))*learn*sum((y-y_pred)*x)
    
y_pred = a + (b*x)
    
plt.scatter(x,y)
plt.plot(x, y_pred, c = 'r')
plt.legend(['y_pred = {}+{}*x'.format(a,b), 'y_actual'],loc='center left', bbox_to_anchor=(1, 0.8))

plt.title("Optimized Fit Line for Data")
plt.show()
```

# Printng Best fit line by Linear Regression model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(x,y)
y_pred = lr.predict(x)
```

```python
c = lr.intercept_
m = lr.coef_
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred, c ='r')
plt.legend(['y_pred = {}+{}*x)'.format(c,m), 'y_actual'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Best Fit Line By Linear Regression Model")
plt.show()
```

# Intution of Gradient Descent
## Seeing How Gradient is converging

```python
a = 0
b = 0

epoch =2000
learn = 0.001

a = np.zeros(epoch+1)
b = np.zeros(epoch+1)

y_pred = a[0] + (b[0]*x)
for i in range(1, epoch+1):
    a[i] = a[i-1] -(-2/len(y))*learn*sum(y-y_pred)
    b[i] = b[i-1] - (-2/len(y))*learn*sum((y-y_pred)*x)
    
    y_pred = a[i] + (b[i]*x)

for i in [0,10,20,50,100,200,500,900,1500,2000]:
    y_predict = a[i]+b[i]*x
    plt.scatter(x,y)
    plt.plot(x, y_predict, c ='r')
    plt.legend(['y_pred = {}+{}*x)'.format(a[i],b[i]), 'y_actual'],loc='center left', bbox_to_anchor=(1, 0.5))
    plt.title('{}-iteration'.format(i))
    plt.show()
```
